## ECRITURE D´UN ALGORITHME

Quand on écrit un algorithme, on utilise un langage  "naturel" ("tant que", "si"...). <br> Ce langage 
appelé aussi **pseudo code** permet de passer facilement à un langage de programmation on dit 
alors que l'on implémente l'algorithme.  <br><br>
Voici un premier algorithme écrit en pseudo code:

```python
DEBUT

    FONCTION RechercheMin(T: tableau de n entiers)
        n ← taille du tableau
        min ← T[0]
        POUR i ALLANT_DE 1 à n
            SI min >T[i]
                min ← T[i]
            FIN SI
        FIN POUR
        RETOURNE min
        
FIN

```
